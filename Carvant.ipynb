```python
#Creating fresh dataframe from Original:

df_color_stock = df_stock
```

```python
#Creating value count of Colors : 

df_color_count = df_stock["CAR_COLOR"].value_counts()

#Reviewing the Color Value Count from top-down:
df_color_count.head(20)
```

```python
#Reviewing the Color Value Count from bottom-up
df_color_count.tail(50)
```

```python
#Prepairing & reviewing data for slicing :
color_slice = df_color_stock.CAR_COLOR.value_counts() 

color_slice
```

```python
#Cleaning Data of colors with value counts of two (2) or less to remove irrelivent deals & entries with spelling mistakes/ unique entries for the color: 

df_color_slice2 = df_color_stock[df_color_stock.CAR_COLOR.isin(color_slice.index[color_slice.gt(2)])]

#Reviewing cleaned data

df_color_slice2
```

```python
#Creating Value count for cleaned data 
color = df_color_slice2.CAR_COLOR.value_counts() 

#Reviewing cleaned data from bottom-up to verify removal of entries. 
color.tail(10)
```

```python
color.head(10)
```

```python
#filtering cleaned data for review
car_color_slice2_bar = df_color_slice2[["CAR_COLOR", "Net CO %"]].groupby("CAR_COLOR").mean("Net CO %").sort_values("CAR_COLOR")

car_color_slice2_bar.append(color, ignore_index=True)

#Reviewing cleaned color data with "Net CO %"
car_color_slice2_bar
```

```python
#Creating & reviewing HVPlot Bar chart of cleaned color data with "Net CO %" data

car_color_slice2_bar.hvplot.bar(width=1500, height=1000, rot=90)
```

```python
# 2. Question 2 answer : Yes, It appears after reviewing the cleaned data that Black has the most losses, as well as representing the most sold overall color. Follwed by "Champaign" & "Burgendy". However "Champaign" & "Burgendy" represent a drasticly lower number of sales than black, and are thous not as statisticly relivant. 
```
